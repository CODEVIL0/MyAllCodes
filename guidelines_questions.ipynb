```python
#Compute mean, standard deviation, and variance of a 2D random integer array along the second axis
import numpy as np

# Create a random 2D array
array = np.random.randint(0, 10, (3, 4))

# Compute mean, standard deviation, and variance along the second axis
mean = np.mean(array, axis=1)
std_dev = np.std(array, axis=1)
variance = np.var(array, axis=1)

print("Mean along second axis:", mean)
print("Standard deviation along second axis:", std_dev)
print("Variance along second axis:", variance)
```

```python
##Test whether the elements of a given 1D array are zero, non-zero and NaN. Record the indices of these elements in three separate arrays
import numpy as np

# Given 1D array
array = np.array([1, 0, 5, np.nan, 3, 0])

# Find indices of zero, non-zero, and NaN elements
zero_indices = np.where(array == 0)[0]
non_zero_indices = np.where(array != 0)[0]
nan_indices = np.where(np.isnan(array))[0]

# Print the indices
print("Indices of Zero Elements:", zero_indices)
print("Indices of Non-Zero Elements:", non_zero_indices)
print("Indices of NaN Elements:", nan_indices)
```

```python
##Create three random arrays of the same size: Array1, Array2 and Array3. Subtract Array 2 from Array3 and store in Array4. Create another array Array5 having two times the values in Array1. Find Covariance and Correlation of Array1 with Array4 and Array5 respectively
import numpy as np

# Create random arrays of the same size
size = 5  # Define the size of the arrays
Array1 = np.random.rand(size)
Array2 = np.random.rand(size)
Array3 = np.random.rand(size)

# Subtract Array2 from Array3 and store in Array4
Array4 = Array3 - Array2

# Create Array5 with values twice those of Array1
Array5 = 2 * Array1

# Find covariance and correlation
covariance = np.cov(Array1, Array4)[0, 1]
correlation = np.corrcoef(Array1, Array5)[0, 1]

print("Array1:", Array1)
print("Array2:", Array2)
print("Array3:", Array3)
print("Array4 (Array3 - Array2):", Array4)
print("Array5 (2 * Array1):", Array5)
print("Covariance of Array1 with Array4:", covariance)
print("Correlation of Array1 with Array5:", correlation)
```

```python
##Create two random arrays of the same size 10: Array1, and Array2. Find the sum of the first half of both the arrays and product of the second half of both the arrays.
import numpy as np

# Create two random arrays of size 10
Array1 = np.random.randint(1, 10, 10)
Array2 = np.random.randint(1, 10, 10)

# Find the sum of the first half of both arrays
sum_first_half = np.sum(Array1[:5]) + np.sum(Array2[:5])

# Find the product of the second half of both arrays
product_second_half = np.prod(Array1[5:]) * np.prod(Array2[5:])

print("Array1:", Array1)
print("Array2:", Array2)
print("Sum of the first half of both arrays:", sum_first_half)
print("Product of the second half of both arrays:", product_second_half)
```

```python
##Create a series with 5 elements, sort on index and values separately
import pandas as pd

# Create a series with 5 elements
series = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# Sort the series on index
sorted_by_index = series.sort_index()
print("Sorted by Index:")
print(sorted_by_index)

# Sort the series on values
sorted_by_values = series.sort_values()
print("\nSorted by Values:")
print(sorted_by_values)
```

```python
##Create a series with N elements with duplicate values, find min and max ranks using 'first' and 'max' methods:
import pandas as pd

# Create a series with duplicate values
series = pd.Series([10, 20, 30, 10, 40, 20, 50])

# Find the minimum and maximum ranks using 'first' and 'max' methods
min_rank_first = series.rank(method='first').min()
max_rank_first = series.rank(method='first').max()

min_rank_max = series.rank(method='max').min()
max_rank_max = series.rank(method='max').max()

print("Minimum rank using 'first' method:", min_rank_first)
print("Maximum rank using 'first' method:", max_rank_first)
print("Minimum rank using 'max' method:", min_rank_max)
print("Maximum rank using 'max' method:", max_rank_max)
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame with 3 columns and 50 rows of random numeric data
np.random.seed(42)
data = np.random.rand(50, 3)
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Replace 10% of values with NaN at random index positions
rows, cols = df.shape
nan_indices = np.random.choice(rows * cols, int(0.1 * rows * cols), replace=False)
df.values.ravel()[nan_indices] = np.nan

# a. Identify and count missing values in the DataFrame
missing_values_count = df.isnull().sum().sum()
print("Missing values count:")
print(missing_values_count)

# b. Drop columns with more than 5 null values
df = df.dropna(thresh=len(df) - 5, axis=1)

# c. Identify the row label with the maximum sum of values and drop that row
max_sum_row_label = df.sum(axis=1).idxmax()
df = df.drop(index=max_sum_row_label)

# d. Sort the DataFrame based on the first column after dropping columns
df = df.sort_values(by=df.columns[0])

# e. Remove duplicates from the first column
df = df.drop_duplicates(subset=df.columns[0])

# Check if DataFrame has at least 2 columns for correlation and covariance calculations
if len(df.columns) >= 2:
    # f. Find correlation between the first and second column, and covariance between the second and third column
    correlation = df[df.columns[0]].corr(df[df.columns[1]])
    covariance = df[df.columns[1]].cov(df[df.columns[2]])
    print("Correlation between first and second column:", correlation)
    print("Covariance between second and third column:", covariance)
else:
    print("Insufficient columns for correlation and covariance calculations.")

# g. Discretize the second column into 5 bins
if 'B' in df.columns:
    df['B_bins'] = pd.cut(df['B'], bins=5)

print("\nFinal DataFrame:")
print(df)
```

```python
import pandas as pd

# Load the attendance data from the two Excel files into separate dataframes
df_workshop1 = pd.read_excel('Workshop1.xlsx')
df_workshop2 = pd.read_excel('Workshop2.xlsx')

# a. Perform merging of the two data frames to find the names of students who attended both workshops
attended_both = pd.merge(df_workshop1, df_workshop2, on='Name', how='inner')['Name']

# b. Find names of all students who have attended a single workshop only
attended_single = pd.concat([df_workshop1, df_workshop2]).drop_duplicates(subset='Name', keep=False)['Name']

# c. Merge two data frames row-wise and find the total number of records in the data frame
merged_df = pd.concat([df_workshop1, df_workshop2], ignore_index=True)
total_records = len(merged_df)

# d. Merge two data frames row-wise and use 'Name' and 'Date' as multi-row indexes
multi_index_df = pd.concat([df_workshop1.set_index(['Name', 'Date']), df_workshop2.set_index(['Name', 'Date'])])

# Generate descriptive statistics for the hierarchical data frame
descriptive_stats = multi_index_df.describe()

# Display the results
print("Names of students who attended both workshops:")
print(attended_both)

print("\nNames of students who attended a single workshop only:")
print(attended_single)

print("\nTotal number of records after merging:")
print(total_records)

print("\nDescriptive statistics for the hierarchical data frame:")
print(descriptive_stats)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map target values to species names
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# a. Plot bar chart to show the frequency of each class label
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df, palette='Set1')
plt.title('Frequency of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

# b. Draw a scatter plot for Petal width vs Sepal width and fit a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='sepal width (cm)', y='petal width (cm)', data=df, scatter=True, fit_reg=True)
plt.title('Petal Width vs Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

# c. Plot density distribution for feature petal length
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='petal length (cm)', shade=True)
plt.title('Density Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.show()

# d. Use a pair plot to show pairwise bivariate distribution
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle('Pairwise Bivariate Distribution')
plt.show()

# e. Draw heatmap for the four numeric attributes
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='YlOrRd')
plt.title('Heatmap of Numeric Attributes')
plt.show()

# f. Compute statistics for each feature
statistics = df.describe()

# g. Compute correlation coefficients between each pair of features and plot heatmap
correlation_matrix = df.drop('species', axis=1).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlOrRd')
plt.title('Correlation Heatmap')
plt.show()

print("Statistics for each feature:")
print(statistics)
```

```python
import pandas as pd

# Create the DataFrame with the provided data
data = {
    'Name': ['Shah', 'Vats', 'Vats', 'Kumar', 'Vats', 'Kumar', 'Shah', 'Shah', 'Kumar', 'Vats'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male'],
    'MonthlyIncome': [114000.00, 65000.00, 43150.00, 69500.00, 155000.00, 103000.00, 55000.00, 112400.00, 81030.00, 71900.00]
}

df = pd.DataFrame(data)

# a. Calculate and display familywise gross monthly income
familywise_income = df.groupby('Name')['MonthlyIncome'].sum()
print("Familywise Gross Monthly Income:")
print(familywise_income)

# b. Calculate and display the member with the highest monthly income
highest_income_member = df.loc[df['MonthlyIncome'].idxmax()]
print("\nMember with the Highest Monthly Income:")
print(highest_income_member)

# c. Calculate and display monthly income of all members with income greater than Rs. 60000.00
high_income_members = df[df['MonthlyIncome'] > 60000.00]
print("\nMonthly Income of Members with Income > Rs. 60000.00:")
print(high_income_members)

# d. Calculate and display the average monthly income of the female members
average_female_income = df[df['Gender'] == 'Female']['MonthlyIncome'].mean()
print("\nAverage Monthly Income of Female Members:")
print(average_female_income)
```

```python
import pandas as pd

# Load the Titanic dataset from an online source
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_data = pd.read_csv(url)

# a. Find total number of passengers with age less than 30
passengers_age_less_than_30 = titanic_data[titanic_data['Age'] < 30]['PassengerId'].count()
print("Total number of passengers with age less than 30:", passengers_age_less_than_30)

# b. Find total fare paid by passengers of first class
total_fare_first_class = titanic_data[titanic_data['Pclass'] == 1]['Fare'].sum()
print("Total fare paid by passengers of first class:", total_fare_first_class)

# c. Compare number of survivors of each passenger class
survivors_by_class = titanic_data.groupby('Pclass')['Survived'].sum()
print("\nNumber of survivors by passenger class:")
print(survivors_by_class)

# d. Compute descriptive statistics for any numeric attribute genderwise
descriptive_stats_genderwise = titanic_data.groupby('Sex').describe()
print("\nDescriptive statistics for numeric attributes genderwise:")
print(descriptive_stats_genderwise)
```
